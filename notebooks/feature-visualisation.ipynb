```python
import src.plotting as plot
from src.helpers import read_video
import matplotlib.pyplot as plt

from src.plotting import plot_continues_line, plot_bodypart_lines, plot_bodypoint, stick_movie, BODYPART_LINES, DEFAULT_BODYPARTS, COLORS 
```

```python
df = read_video('001')
```

```python
fig = plt.figure(figsize=(12,14))

# plot the lines (based on average values)
plot_bodypart_lines(df)

# give the wrists (based on average values) their own color 
plot_bodypoint(df, 'wrist1', 'bo', alpha=1)
plot_bodypoint(df, 'wrist2', 'bo', alpha=1)

# plot all bodyparts from whole video
for bodypart, color in zip(DEFAULT_BODYPARTS, COLORS+COLORS):
    for i in range(len(df)):
        plot_bodypoint(df, bodypart, color, frame=i)
        
# plot continues lines between shoulders and hips
plot_continues_line(df, 'shoulder1', 'hip1')
plot_continues_line(df, 'shoulder2', 'hip2')

# invert axis
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
```
